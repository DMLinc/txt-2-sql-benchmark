```python
from typing import Tuple

READONLY_CON = con_viewer

def try_run_query(sql: str) -> pd.DataFrame | str:
    try:
        return pd.read_sql_query(sql, con=READONLY_CON)
    except Exception as e:
        print('failed to execute')
        return e

def compare_queries(benchmark_sql: str, submission_sql: str) -> dict[str, bool|str]:
    benchmark_loc_df = try_run_query(benchmark_sql)
    submission_loc_df = try_run_query(submission_sql)
    # print(submission_loc_df)
    # print(isinstance(submission_loc_df, pd.DataFrame))
    if isinstance(submission_loc_df, pd.DataFrame):
        res = compare_results(
            df_true=benchmark_loc_df,
            df_pred=submission_loc_df
        )
        if res:
            return {'score': res, 'reason': 'No error'}
        else:
            return {'score': res, 'reason': 'Incorrect output'}
    else:
        return {'score': False, 'reason': f"Failed to execute: {submission_loc_df}"}

def execution_accuracy(benchmark: pd.DataFrame, submission_df: pd.DataFrame):
    merged = pd.merge(benchmark, submission_df, on='question', how='left')

    results = []
    for i, row in merged.iterrows():
        # print(row['correct_sql_query'])
        # print(row['submitted_sql'])
        score_dict = compare_queries(row['correct_sql_query'], row['submitted_sql'])
        score_dict['question'] = row['question']
        results.append(score_dict)
    
    return pd.merge(submission_df, pd.DataFrame(results), on='question')
```
